## Introduction

The World Health Organization recommends that adults undertake at least 150 minutes of ‘moderate-intensity aerobic physical activity’ per week (from now on referred to as ‘exercise’). Adults who do not get this amount are classed as inactive (Physical activity, 2020). 

In 2018, approximately 27.5% of all adults worldwide were inactive. Figures show that women tend to get less exercise than men. The global average for inactive women in 2019 was 31.7%, compared with 23.4% for men (GHO | By category | Prevalence of insufficient physical activity among adults - Data by WHO region, 2018). Inactivity also tends to increase with age, with over 60% of 60+ year olds in the Americas being inactive in 2011 (Hallal et al., 2012).

The Covid-19 pandemic has dramatically changed the way we live, from working practices to leisure activities and even mental wellbeing affected. An important consideration is how the pandemic has impacted exercise. Early on, sales of exercise equipment increased significantly, with many shops unable to keep up with demand (Goldbaum, 2020; Fenton, 2020). Meanwhile, with social-distancing paramount, the majority of team sports were either suspended or cancelled, at both professional and grassroots level (BBC Sport, 2020). 

This report aims to examine the impact in more detail. Is the new equipment being used, or merely taking up space? Has the cessation of team sports stopped participants from exercising entirely, or have they found new ways to exercise? Has the gender/age gap changed? In answering these questions, this report aims to present a clear picture of the impact of the pandemic. With this information, it is hoped that the long-term exercise outlook can be improved. If exercise levels have increased, reasons should be established in order to make recommendations or even societal changes to ensure that levels remain high once the pandemic is over. Conversely, if levels have decreased, further research should address what can be done to increase physical activity while the pandemic continues.

The population for this project is all adults (aged 18 and over) who are capable of some form of physical exercise, and who reside in a country which has reported cases of Covid-19. The data has been gathered using opportunity sampling due to time constraints. The sample may display some bias for two reasons. Firstly, the majority of responses come from one age range. Secondly, the responses are mainly from affluent countries. Countries in the World Bank’s high-income group show the highest levels of inactivity, at 36.8% overall (GHO | By category | Prevalence of insufficient physical activity among adults - Data by World Bank income groups, 2018). If the sample population contained more participants from middle and low-income countries, the results may be quite different.

To address the objectives of the project, the research question is framed as follows: Has the Covid-19 pandemic affected people’s exercise habits? Supplementary questions include how habits have changed, and whether there is a relationship between demographic factors (age, gender, and marital status) and the changes resulting from the pandemic. While it may have been interesting and informative to also consider other factors, such as geographical or socio-economic factors, for reasons of practicality these factors are not considered here.

The terms used throughout the report are well-known and require no definition. 

## Survey

The survey consists of three demographic questions and nine topic-specific questions. The demographic questions are at the beginning of the survey as these are generally easier. The gender question includes a ‘prefer not to say’ option as this can be sensitive. Similarly, age is divided into ranges.

The topic-specific questions follow on a new page, to distinguish them from the demographic questions. The most important appear first, with the attitude-based questions last. If a respondent ends the survey early, the most important information should have been recorded. 

The questions and responses have been worded to avoid bias or presumptions, for example including a 0 option for number of hours, rather than presuming all participants do some exercise (although see discussion section for comment). The ‘amount of exercise’ responses are expressed numerically, rather than using ambiguous terms such as ‘a lot’, and the attitude questions use Likert-scales to capture consistent, ordinal data.

‘Type’ and ‘reason’ questions include a free form text box, given that there are a multitude of exercises and motivations. Although coding and analysing such responses is more involved, it is considered necessary for the accuracy of the research.

## Discussion

From the analysis, it can certainly be concluded that the Covid-19 pandemic has affected people’s exercise habits. Types of exercise have changed, with a reduction in team sports (no doubt a result of social distancing) and swimming (likely due to a reluctance/inability to visit indoor swimming pools).

Mental wellbeing has become the most important reason to exercise. This supports the contention that Covid-19 is having a significant impact on people’s mental health, and that exercise can help with this. This could be an area for governments and organisations to focus on in the future, rather than the more traditional health/appearance benefits.

There has been a change both in who people exercise with and where they exercise. Again, this is likely due to social distancing, ‘bubbles’ and closures of indoors venues. There has been a number of global initiatives aimed at promoting exercises at home during lockdown (#HealthyAtHome - Physical activity, 2020),  and this research suggests that these initiatives have been successful.

Rather than just closing, the gender gap appears to have actually reversed. This may be due to team sports stopping (82% of team sports were done by males pre-pandemic) and gyms closing. However, the increase in females exercising is certainly positive, and their primary reason for exercising lends further support to the idea of promoting exercise for mental health reasons.

Other conclusions should be drawn with caution. While it may well be correct that the variety of exercises performed reduces with age, it is noted that the number of responses for the higher age groups is much lower than lower age groups. Similarly, while results show people aged 60-79 doing on average 11 hours per week during the pandemic, this is based on just nine responses. More data would be required for older people in order to reach firm conclusions. Likewise, with only one divorced respondent, forming valid conclusions based on marital status is impossible.  

There are also flaws in the research that should be addressed. Firstly, converting age groups and hours to integers is imprecise. This is especially so for hours, given that the upper group is open ended. While it is unlikely that many people are exercising for substantially more than 15 hours, if the project were repeated, integer values would be required for hours and ages.

Regarding the survey, the first option for amount of exercise is 0-3 hours. In retrospect, 0 hours should have been a separate option, as no exercise is very different from any exercise. Furthermore, 150 minutes should have been the first upper boundary to distinguish between active and inactive. Secondly, the follow-up questions should have included a ‘none’ option for those who did no exercise.

Further research should focus on whether exercise has actually succeeded in improving mental health, and whether this applies equally across genders and age groups. If it has helped, the link should be further developed so that inactivity continues to decrease once the pandemic is over.

```python
# Function to display ages as bar chart and pie chart side by side
def age_func(df):
    fig = plt.figure(figsize=(12,6))
    fig.suptitle('Ages of Respondents',fontsize=20)
    age_counts = df['Age'].value_counts()
    age_counts.sort_index(inplace=True)
    ax1 = fig.add_subplot(1,2,1)
    age_counts.plot.bar(ax=ax1)
    ax1.set_xlabel ('Age', fontsize=12)
    ax1.set_ylabel ('Total', fontsize=12)
    ax1.spines["top"].set_visible(False) # Preferred for aesthetics
    ax1.spines["right"].set_visible(False) # Preferred for aesthetics
    plt.xticks(rotation=45) # Easier to read
    ax2 = fig.add_subplot(1,2,2)
    age_counts.plot.pie(ax=ax2, explode=(0.2, 0.1, 0.2, 0.2, 0.2, 0.2), # Easier to see
                         autopct='%1.0f%%', # Include percentage
                         counterclock=False,
                         shadow = True,
                         startangle=120,
                        labels = None, # Remove segment labels
                        label='') # Remove y label
    ax2.legend(age_counts.index, title="Ages",
         bbox_to_anchor=(0.75, 0, 0.5, 1))
    plt.show()
```

```python
# Function to plot hours per for two genders both before and during covid
def gender_hours(df1,df2):
    fig = plt.figure(figsize=(15,12))
    fig.suptitle('Hours of Exercise per Week by Gender',fontsize=20)
    ax1 = fig.add_subplot(2,2,1)
    sns.histplot(df1[df1['Gender']=='Female']['Hours per week pre-Covid'], bins = 6,
                 kde = True, ax=ax1, color='tab:red') # Use different colour to better distinguish genders
    ax1.axvline(x=df1[df1['Gender']=='Female']['Hours per week pre-Covid'].mean(),
                color='k', linestyle = 'dashed', linewidth = 1)
    ax1.annotate('skew = {}'.format(df1[df1['Gender']=='Female']['Hours per week pre-Covid'].skew().round(2)),
                 xy=(13,15),fontsize=15)
    ax1.annotate('mean', xy=(df1[df1['Gender']=='Female']['Hours per week pre-Covid'].mean(), 15),fontsize=15,
                 xytext=(2, 15), arrowprops=dict(arrowstyle='->'))
    ax1.set_xlabel ('Hours per week (Total hours: {})'.format
                    (df1[df1['Gender']=='Female']['Hours per week pre-Covid'].sum()),fontsize=11)
    ax1.set_ylabel ('Total', fontsize=11)
    ax1.set_title('Female - Pre-Covid-19 Pandemic', fontsize=16, pad=25)
    plt.xticks([2.5,5,7.5,10,12.5,15],
           ["0-3", "3-6", "6-9", "9-12", "12-15", "15+"])
    ax2 = fig.add_subplot(2,2,2)
    sns.histplot(df1[df1['Gender']=='Male']['Hours per week pre-Covid'], bins = 6, kde = True, ax=ax2)
    ax2.axvline(x=df1[df1['Gender']=='Male']['Hours per week pre-Covid'].mean(),
                color='k', linestyle = 'dashed', linewidth = 1)
    ax2.annotate('skew = {}'.format(df1[df1['Gender']=='Male']['Hours per week pre-Covid'].skew().round(2)),
                 xy=(13,15),fontsize=15)
    ax2.annotate('mean', xy=(df1[df1['Gender']=='Male']['Hours per week pre-Covid'].mean(), 15),fontsize=15,
                 xytext=(2, 15), arrowprops=dict(arrowstyle='->'))
    ax2.set_xlabel ('Hours per week (Total hours: {})'.format
                    (df1[df1['Gender']=='Male']['Hours per week pre-Covid'].sum()),fontsize=11)
    ax2.set_ylabel ('')
    ax2.set_title('Male - Pre-Covid-19 Pandemic', fontsize=16, pad=25)
    plt.ylim(0,17)
    plt.xticks([2.5,5,7.5,10,12.5,15],
           ["0-3", "3-6", "6-9", "9-12", "12-15", "15+"])
    ax3 = fig.add_subplot(2,2,3)
    sns.histplot(df2[df2['Gender']=='Female']['Hours per week during Covid'], bins = 6,
                 kde = True, ax=ax3, color='tab:red')
    ax3.axvline(x=df2[df2['Gender']=='Female']['Hours per week during Covid'].mean(),
                color='k', linestyle = 'dashed', linewidth = 1)
    ax3.annotate('skew = {}'.format(df2[df2['Gender']=='Female']['Hours per week during Covid'].skew().round(2)),
                 xy=(13,15),fontsize=15)
    ax3.annotate('mean', xy=(df2[df2['Gender']=='Female']['Hours per week during Covid'].mean(), 15),fontsize=15,
                 xytext=(2, 15), arrowprops=dict(arrowstyle='->'))
    ax3.set_xlabel ('Hours per week (Total hours: {})'.format
                    (df2[df2['Gender']=='Female']['Hours per week during Covid'].sum()),fontsize=11)
    ax3.set_ylabel ('Total', fontsize=11)
    ax3.set_title('Female - During Covid-19 Pandemic', fontsize=16, pad=25)
    plt.ylim(0,17)
    plt.xticks([2.5,5,7.5,10,12.5,15],
           ["0-3", "3-6", "6-9", "9-12", "12-15", "15+"])
    ax4 = fig.add_subplot(2,2,4)
    sns.histplot(df2[df2['Gender']=='Male']['Hours per week during Covid'], bins = 6, kde = True, ax=ax4)
    ax4.axvline(x=df2[df2['Gender']=='Male']['Hours per week during Covid'].mean(),
                color='k', linestyle = 'dashed', linewidth = 1)
    ax4.annotate('skew = {}'.format(df2[df2['Gender']=='Male']['Hours per week during Covid'].skew().round(2)),
                 xy=(13,15),fontsize=15)
    ax4.annotate('mean', xy=(df2[df2['Gender']=='Male']['Hours per week during Covid'].mean(), 15),fontsize=15,
                 xytext=(2, 15), arrowprops=dict(arrowstyle='->'))
    ax4.set_xlabel ('Hours per week (Total hours: {})'.format
                    (df2[df2['Gender']=='Male']['Hours per week during Covid'].sum()),fontsize=11)
    ax4.set_ylabel ('')
    ax4.set_title('Male - During Covid-19 Pandemic', fontsize=16, pad=25)
    plt.ylim(0,17)
    plt.xticks([2.5,5,7.5,10,12.5,15],
           ["0-3", "3-6", "6-9", "9-12", "12-15", "15+"])
    plt.tight_layout()
    plt.show()
```
